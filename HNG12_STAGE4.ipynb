```python
import os

def find_files_in_path(path, file_extension=None):

  file_paths = []
  for root, _, files in os.walk(path):
    for file in files:
      if file_extension is None or file.endswith(file_extension):
        file_paths.append(os.path.join(root, file))
  return file_paths

# Example usage:
path_to_search = '/root/.cache/kagglehub/datasets/gauthamp10/apple-appstore-apps/versions/1'
csv_files = find_files_in_path(path_to_search, file_extension='.csv')
print(csv_files)
```
